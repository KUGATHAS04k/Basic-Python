##Brodcasting using numPY


*   They are equal
*   one of them is 1.

### The Step-by-Step Process
* Align Dimensions: Align the shapes of the two arrays at the right.


* Check Compatibility: * If the dimensions are the same, you're good.

If one dimension is 1, that array is "stretched" to match the other.

If one array has fewer dimensions, prepended "1"s are added to its shape on the left until the ranks match.

* Fail Condition: If at any point the dimensions are different and neither is 1, NumPy raises a ValueError.



```python
array5 = np.array([[1,2,3,4],[4,5,6,7],[7,8,9,4],[1,5,9,8]])
array6 = np.array([[1],[2],[3],[4]])


print(array5.shape)
print(array6.shape)
```

```python
print(array5 * array6)
```
